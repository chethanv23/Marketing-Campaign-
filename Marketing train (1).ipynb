# Outliers Analysis

```python
Numer = [feature for feature in data.columns if data[feature].dtypes != 'O']
```

```python
data[Numer]
```

```python
Numer
```

```python
Col_dic = {
    'campaign':1,
 'pdays':2,
 'previous':3,
 'emp.var.rate':4,
 'cons.price.idx':5,
 'cons.conf.idx':6,
 'euribor3m':7,
 'nr.employed':8,
 'pmonths':9,
 'pastEmail':10,
'custAge':11}

mlp.figure(figsize=(20 , 30))

for variable,i in Col_dic.items():
    mlp.subplot(3,4,i)
    mlp.boxplot(data[variable] , whis= 1.5)
    mlp.title(variable)
```

```python
for variable in Col_dic:
    q25,q75 = np.percentile(data[variable],[25,75])
    iqr = q75-q25
    
    min_val = q25 - (1.5*iqr)
    max_val = q75 + (1.5*iqr)
    
    print("number of outliers present in the variable {} is {} and the percentage of ouliers is {}" .format(variable , len((np.where((data[variable] > max_val) | 
                                                                                           (data[variable] < min_val))[0])),len((np.where((data[variable]  > max_val) | 
                                                                                           (data[variable]  < min_val))[0]))*100/len(data)))
```

From the box plot and the turky's method we can have the following observations.
    
    1)There are four variables , which got no outliers
            * emp.var.rate
            * cons.price.idx
            * euribor3m
            * nr.employed
    2)For those variables which gogot outliers ,let's replace the outliers with winsorise technique


```python
from scipy.stats.mstats import winsorize
data['campaign'] = winsorize(data['campaign'] , (0,0.06))
data['pdays'] = winsorize(data['pdays'] , (0.04,0))
data['previous'] = winsorize(data['previous'] , (0,0.18))
data['cons.conf.idx'] = winsorize(data['cons.conf.idx'] ,(0,0.02))
data['pmonths'] = winsorize(data['pmonths'], (0.04 , 0))
data['pastEmail'] = winsorize(data['pastEmail'] , (0.13 , 0.13))
data['custAge'] = winsorize(data['custAge'] ,(0.01 , 0.04))
```

```python
Col_dic = {
    'campaign':1,
 'pdays':2,
 'previous':3,
 'emp.var.rate':4,
 'cons.price.idx':5,
 'cons.conf.idx':6,
 'euribor3m':7,
 'nr.employed':8,
 'pmonths':9,
 'pastEmail':10,
'custAge':11}

mlp.figure(figsize=(20 , 30))

for variable,i in Col_dic.items():
    mlp.subplot(3,4,i)
    mlp.boxplot(data[variable] , whis= 1.5)
    mlp.title(variable)
```

```python
for variable in Col_dic:
    q25,q75 = np.percentile(data[variable],[25,75])
    iqr = q75-q25
    
    min_val = q25 - (1.5*iqr)
    max_val = q75 + (1.5*iqr)
    
    print("number of outliers present in the variable {} is {} and the percentage of ouliers is {}" .format(variable , len((np.where((data[variable] > max_val) | 
                                                                                           (data[variable] < min_val))[0])),len((np.where((data[variable]  > max_val) | 
                                                                                           (data[variable]  < min_val))[0]))*100/len(data)))
```

Observation made from the heatmap

    *emp.ar.rate is highly corelated with the many variables so we can take out the variable in feature selection

Simillarly ,To reduce the dimensionality of the categorical variable we need build the chi_square contingency table. According to the p value( significance value ,usualy 0.05 )we reject alternate hypothesis or null hypothesis

```python
from scipy.stats import chi2_contingency
for i in cat_var:
    print(i)
    chi2, p , dof,ex = chi2_contingency(pd.crosstab(data['responded'] , data[i]))
    print(p)
```

from the p value we got in the above chi_square test. for variables whose p value is greater than 0.05 significance value then we reject the null hypothesis(that the two variables are dependent)

# Model development

```python
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
data1['responded'] = data1['responded'].replace(0 , 'No')
data1['responded'] = data1['responded'].replace(1 , 'Yes')
```

```python
data1.head(20)
```

```python
X = data1.drop(['responded'] , axis = 1)
Y = data1['responded']
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, Y,  stratify= Y,test_size = 0.2)
```

```python
print(X_train.shape , X_test.shape , y_train.shape , y_test.shape)
```

```python
C50_model = tree.DecisionTreeClassifier(criterion='entropy').fit(X_train , y_train)
```

```python
C50_Predictions = C50_model.predict(X_test)
```

```python
C50_Predictions
```

```python
dotfile = open("pt.dot" , 'w')
df = tree.export_graphviz(C50_model , out_file=dotfile , feature_names= X_train.columns)
```
